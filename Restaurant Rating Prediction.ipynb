# import necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py
import seaborn as sns
```

```python
data = pd.read_csv("zomato.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
#checking the data types

data.dtypes
```

```python
#checking null values

data.isna().sum()
```

```python
#Deleting Unnnecessary Columns
#Dropping the column like "phone" and "url" and saving the new dataset as "df"

df=data.drop(['url','phone'],axis=1) 
```

# checking for duplicated value

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# Drop null value

```python
#Remove the NaN values from the dataset

df.dropna(how='any',inplace=True)
df.isnull().sum()
```

# Rename columns

```python
df.columns
```

```python
df = df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type',
                                  'listed_in(city)':'city'})
df.columns
```

```python
df.head()
```

# cleaning the dataset

```python
df['cost'].unique()
```

# replacing "," with nothing and converting the result to float

```python
#zomato['cost'] = zomato['cost'].astype(str) #Changing the cost to string
#Using lambda function to replace ',' from cost

df['cost'] = df['cost'].apply(lambda x: x.replace(',','')) 
df['cost'] = df['cost'].astype(float)
```

```python
print(df['cost'].unique())

print('---'*10)

df.dtypes
```

```python
#Reading uninque values from the Rate column

df['rate'].unique()
```

```python
#getting rid of "NEW"

df = df.loc[df.rate !='NEW'] 
```

```python
df['rate'].unique()
```

```python
#Removing '/5' from Rates

df['rate'] = df['rate'].apply(lambda x: x.replace('/5',''))
```

```python
x = df.iloc[:,[2,3,5,6,7,9,10,12]]
x.head()
```

```python
y = df['rate']
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import  ExtraTreesRegressor
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=10)
```

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=lr_model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_Model=RandomForestRegressor(n_estimators=650,random_state=245,min_samples_leaf=.0001)
RF_Model.fit(x_train,y_train)
y_predict=RF_Model.predict(x_test)
r2_score(y_test,y_predict)
```

```python
#Preparing Extra Tree Regression
from sklearn.ensemble import  ExtraTreesRegressor
ET_Model=ExtraTreesRegressor(n_estimators = 120)
ET_Model.fit(x_train,y_train)
y_predict=ET_Model.predict(x_test)


from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

```python
#Use pickle to save our model so that we can use it later

import pickle 
# Saving model to disk
pickle.dump(ET_Model, open('model.pkl','wb'))
model=pickle.load(open('model.pkl','rb'))
```
